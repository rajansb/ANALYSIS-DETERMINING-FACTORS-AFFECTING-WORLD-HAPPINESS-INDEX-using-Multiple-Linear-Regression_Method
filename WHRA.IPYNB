```python
from sklearn.linear_model import LinearRegression # Regression models
```

```python
def linear(var):
    y = df['Happiness_Score']
    z = df[[var]]
    lm1 = LinearRegression()
    lm1.fit(z,y)
    return lm1.coef_,lm1.intercept_,lm1.score(z,y),(np.sqrt((y-lm1.predict(z))**2/(y.count()-2))).mean()
```
